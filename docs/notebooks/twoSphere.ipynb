```python
from dreimac import GeometryExamples, PlotUtils, ComplexProjectiveCoords, ProjectiveMapUtils, GeometryUtils

import numpy as np
from persim import plot_diagrams
```

```python
data = GeometryExamples.sphere(2000)

#%matplotlib widget
#from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(data[:,0], data[:,1], data[:,2],s=20, c=data[:,2])
```

```python
data_R2 = ProjectiveMapUtils.stereographic_projection_hemispheres(data)
fig = plt.figure()
plt.scatter(data_R2[:,0], data_R2[:,1],c=data[:,2])
_ = plt.axis("equal")
```

```python
cpc = ComplexProjectiveCoords(data, n_landmarks=100)

fig = plt.figure()
plot_diagrams(cpc.dgms_)
```

```python
coords = cpc.get_coordinates(perc=0.8)
print(coords.shape)
coords_R3 = ProjectiveMapUtils.hopf_map(coords)
```

```python
coords_R2 = ProjectiveMapUtils.stereographic_projection_hemispheres(coords_R3, center_vector=np.array([1,0,0.7]))
print(coords_R2.shape)
fig = plt.figure()
plt.scatter(coords_R2[:,0], coords_R2[:,1],c=data[:,2])
_ = plt.axis("equal")
```
